```python
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
import json
import boto3
import csv
```

```python
#opening pdp sitemap to get product links and then making a pandas dataframe from it
def sitemap_to_dataframe(url):
  response = urllib.request.urlopen(url)
  xml = BeautifulSoup(response, 'lxml-xml', from_encoding=response.info().get_param('charset'))
  df = pd.DataFrame(columns=['loc', 'name', 'productcode', 'detail', 'price', 'stock'])
  urls = xml.find_all("url")
  for url in urls:
    if url.find("loc"):
          loc = url.find("loc").text
          try:
            page = urllib.request.urlopen(loc)
            soup = BeautifulSoup(page, 'html.parser')
            name_box = soup.find('h1', attrs={'class':'product__name'})
            detail_box = soup.find('div', attrs={'id': 'product-description'})
            #checking the script in page for price or out of stock
            price_box = soup.find('script', attrs={'id':'__NEXT_DATA__'})
            if name_box and detail_box and price_box:
              name_box = name_box.text.strip()
              detail_box = detail_box.text.strip()
              price_box = price_box.text.strip()
              price_box = json.loads(price_box)
              productcode_box = list(price_box['props']['initialState']['products']['productsByProductCode'].keys())[0]
              stock_box = price_box['props']['initialState']['products']['productsByProductCode'][productcode_box]['isProductOutOfStock']
              if stock_box:
                price_box = 'OUT OF STOCK'
                stock_box = 'No'
              else:
                price_box = price_box['props']['initialState']['products']['productsByProductCode'][productcode_box]['priceModel'][productcode_box]['colors'][0]['salePrice']['formatted']
                stock_box = 'Yes'
            else:
              name_box = ''
              productcode_box = ''
              detail_box = ''
              price_box = ''
              stock_box = ''
            df = df.append({'loc': loc,'name': name_box,'productcode': productcode_box, 'detail': detail_box, 'price': price_box, 'stock': stock_box}, ignore_index=True)

          except urllib.error.HTTPError as e:
            print(f"HTTPError: {e.code} - {e.reason}")
            loc = ''
            name_box = ''
            productcode_box = ''
            detail_box = ''
            price_box = ''
            stock_box = ''
    else:
          loc = ''
          name_box = ''
          productcode_box = ''
          detail_box = ''
          price_box = ''
          stock_box = ''
  return df
```

```python
url = "https://www.jcrew.com/sitemap-wex/sitemap-pdp1.xml"
df = sitemap_to_dataframe(url)
```
